```python
def score_submission(fpath: str):
    
    print("Loading", fpath)
    # Submission might have been uploaded as .zip or .tar.gz butKaggle renamed all submission to .csv
    # therefore if loading fails, compressed formats are checked
    try:
        submission = pl.read_csv(fpath)
    except:
        try:
            print("extracting file from zip")
            with zipfile.ZipFile(fpath, 'r') as zip_ref:
                zip_ref.extractall(fpath.rsplit('/', 1)[0])
            filenames = glob.glob(fpath.rsplit('/', 1)[0]+"/*")
            filenames.remove(fpath)
            submission = pl.read_csv(filenames[0])
        except:
            #print("extracting file from tar")
            #with tarfile.open(fpath, 'r') as tar:
            #    tar.extractall(fpath.rsplit('/', 1)[0])
            #filenames = glob.glob(fpath.rsplit('/', 1)[0]+"/*")
            #filenames.remove(fpath)
            #submission = pl.read_csv(filenames[0])
            
            print(fpath, 'has unknown filetype')
            
    print(f"shape={submission.shape}")
    df = solution.join(submission.rename({'target': 'prediction'}), on='row_id', how='inner')
    public = df.filter(pl.col('Usage') == 'Public').drop('Usage')
    private = df.filter(pl.col('Usage') == 'Private').drop('Usage')
    
    public_corrs = public.groupby('group_id').agg(pl.corr('prediction', 'target'))
    private_corrs = private.groupby('group_id').agg(pl.corr('prediction', 'target'))
    
    print("public_corrs.shape: ", public_corrs.shape)
    print("private_corrs.shape: ", private_corrs.shape)

    
    public_corrs = public_corrs.to_pandas()
    public_corrs = public_corrs.merge(solution[["group_id","technology"]].to_pandas().drop_duplicates(),on="group_id",how="left")
    
    private_corrs = private_corrs.to_pandas()
    private_corrs = private_corrs.merge(solution[["group_id","technology"]].to_pandas().drop_duplicates(),on="group_id",how="left")
    
    df_results = public_corrs.groupby("technology").prediction.mean().to_frame("public").merge(private_corrs.groupby("technology").prediction.mean()\
                                                                              .to_frame("private"),
                                                                              left_index=True,right_index=True)
    df_results.loc["total_score","public"] = public_corrs.prediction.mean()
    df_results.loc["total_score","private"] = private_corrs.prediction.mean()

    return df_results
```
